```python
import itertools
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
e_i = np.array(list(itertools.product([-1,0,1],[-1,1],[-1,1],[-2,-1,1,2])))
e_i.shape
```

```python
data = pd.DataFrame(e_i,columns = ['tipo_suspension','abertura_malla','temp_ciclaje','replica'])
data.head(10)
```

```python
y = np.array([60,75,75, 86, 67,73,73, 67, 62,68,65, 76, 71,80,80, 80, 76,71,75, 70,75,75,75,77,55,53,53, 55,52,52,57, 54, 44,44,45, 48, 60,60,60,67, 52, 51, 50, 52, 56,55,57,59])
y
```

```python
df_y = pd.DataFrame(y, columns =['vol_sedimentacion'])
df_y.head()
```

```python
df = pd.concat([data, df_y], axis = 1)
df.head(10)
```

```python
df.describe()
```

```python
model = sm.OLS(df['vol_sedimentacion'], df.drop('vol_sedimentacion',axis = 1)).fit()
```

```python
print(model.summary())
```
